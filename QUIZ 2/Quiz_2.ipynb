## **Reading Image**

```python
import cv2 #import opencv library
import numpy as np #import numpy library
from matplotlib import pyplot as plt #import matplotlib library

img = cv2.imread('lego.png', cv2.IMREAD_GRAYSCALE) #reads the image lego.png in grayscale
_, mask = cv2.threshold(img, 150, 250, cv2.THRESH_BINARY_INV) #converts the image into binary format
kernal = np.ones((9,10), np.uint8) #kernal is initialised to move over the image
plt.imshow(mask, 'gray') #shows the image in binary form
```

## **Morphology**
Morphological transformations are some simple operations based on the image shape. It is normally performed on binary images. It needs two inputs, one is our original image, second one is called structuring element or kernel which decides the nature of operation. Two basic morphological operators are Erosion and Dilation. Then its variant forms like Opening, Closing, Gradient etc also comes into play.

```python
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal) #It is useful in closing small holes inside the foreground objects, or small black points on the object.
plt.imshow(closing, 'gray')
```

## **Contour**
Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

In OpenCV, finding contours is like finding white object from black background. Object to be found should be white and background should be black.

```python
(cnt, heirarchy) = cv2.findContours(closing.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # There are three arguments in cv2.findContours() function, first one is source image, second is contour retrieval mode, third is contour approximation method. And it outputs the contours and hierarchy. 
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Converts image from BGR color space to RGB.
cv2.drawContours(rgb, cnt, -1, (150,255,0), 2) # Draws the contours around the objects

plt.imshow(rgb)
     
```

```python
print('Number of object ', len(cnt)) # Counts the number of objects in the image
plt.show()
```
